Using simple few lines of coding to extract list of information from HTML webpage.

Objectives: To scrap company information (name, description, website) from HTML webpage about the list of 68 Top Indonesia Automotive Companies and Startups of 2021 from the website https://beststartup.asia/68-top-indonesia-automotive-companies-and-startups-of-2021/.

Library: Beautiful Soup 4

## Importing modules
**[Request](https://requests.readthedocs.io/en/master/)** Module is used to get the HTML code for the URL given.

**Note**: *Not all webpages can be requested. For example most social media does not allow to scrape data due to privacy issues. These pages require special access of Developer APIs to scrape data.*

```python
import requests 
from bs4 import BeautifulSoup 
from tqdm import tqdm
```

## Define URL

```python
URL = "https://beststartup.asia/68-top-indonesia-automotive-companies-and-startups-of-2021/"
```

### Requesting desired Webpage

```python
r = requests.get(URL)    
soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify()[6000:7000])
```

To extract data: Use ctrl+shift+c to extract element.

element.png

```python
# Extracting all companies info
## Note col._2wzgFH.K0kLPL means 3 entities namely 'col', ' _2wzgFH' and 'K0kLPL' 
## This is written in HTML as 'col _2wzgFH K0kLPL'
## This can also be seen in Bullet 3

row = soup.find_all('div',attrs={"class":"wp-block-cover__inner-container"})
```

To extract company website links, company names, and descriptions.

```python
# use text='Website' to locate the company website link
for x in row:
  print (x.find('a', text='Website').get("href"))
```

```python
for x in row:
  print(x.find('h3').text)
```

The description is at the second 'p'.

```python
for x in row:
  print(x.find('p').findNext("p").get_text())
```

## To store in the dataframe

```python
# list to store data
dataset = []

# iteration over all blocks
for x in row: 
            
    # extracting text from 1st and 2nd row
    Company = x.find('h3').text
    Description = x.find('p').findNext("p").get_text()
    Website = x.find('a', text='Website').get("href")
    
    # appending to data
    dataset.append({'Company': Company , 'Description' : Description, 'Website' : Website})

dataset[:5]
```

The company list was extracted from the website with company name, description, and website link.

## Export Data

```python
# Convert dataframe to pandas dataframe
import pandas as pd
df = pd.DataFrame(data=dataset)
```

Export as csv 

```python
from google.colab import files

df.to_csv('Indonesia Automotive Companies.csv')
files.download('Indonesia Automotive Companies.csv')
```
